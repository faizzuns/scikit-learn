# 2. Pembelajaran 90% split trains dan 10% test

```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()
```

```python
from sklearn.metrics import confusion_matrix

X_train, X_test, y_train, y_test = model_selection.train_test_split( iris['data'], iris['target'], train_size=0.9, test_size=0.1)
print("X trains = ", X_train)
print("X test", X_test)
print("y train", y_train)
print("y test", y_test)
```

## Naive Bayes

```python
iris_gnb_train = gnb.fit(X_train, y_train)

gnb_pred = iris_gnb_train.predict(X_test)

gnb_acc = metrics.accuracy_score(y_test, gnb_pred) * 100
gnb_prec = metrics.precision_score(y_test, gnb_pred, average='macro') * 100
gnb_rec = metrics.recall_score(y_test, gnb_pred, average='macro') * 100
print('Akurasi Naive Bayes =', gnb_acc, '%')
print('Presisi Naive Bayes =', gnb_prec, '%')
print('Recall Naive Bayes =', gnb_rec, '%', '\n')

cnf_matrix_gnb = confusion_matrix(y_test, gnb_pred)
np.set_printoptions(precision=2)

# Plot non-normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix_gnb, classes=iris.target_names,
                      title='Naive Bayes\' Confusion matrix, without normalization')

# Plot normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix_gnb, classes=iris.target_names, normalize=True,
                      title='Naive Bayes\' Normalized confusion matrix')

plt.show()
```

## Decision Tree

```python
iris_dtl_train = dtl.fit(X_train, y_train)

dtl_pred = iris_dtl.predict(X_test)

dtl_acc = metrics.accuracy_score(y_test, dtl_pred) * 100
dtl_prec = metrics.precision_score(y_test, dtl_pred, average='macro') * 100
dtl_rec = metrics.recall_score(y_test, dtl_pred, average='macro') * 100
print('Akurasi Decision Tree =', dtl_acc, '%')
print('Presisi Decision Tree =', dtl_prec, '%')
print('Recall Decision Tree =', dtl_rec, '%', '\n')

cnf_matrix_dtl = confusion_matrix(y_test, gnb_pred)
np.set_printoptions(precision=2)

# Plot non-normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix_dtl, classes=iris.target_names,
                      title='Decision Tree Learning\'s Confusion matrix, without normalization')

# Plot normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix_dtl, classes=iris.target_names, normalize=True,
                      title='Decision Tree Learning\'s Normalized confusion matrix')

plt.show()
```

#  k-Nearest Neighbors

```python
iris_knn_train = neighbors.nearest_centroid.NearestCentroid().fit(X_train, y_train)
knn_pred = iris_knn_train.predict(X_test)

knn_acc = metrics.accuracy_score(y_test, knn_pred) * 100
knn_prec = metrics.precision_score(y_test, knn_pred, average='macro') * 100
knn_rec = metrics.recall_score(y_test, knn_pred, average='macro') * 100
print('Akurasi KNN =', knn_acc, '%')
print('Presisi KNN =', knn_prec, '%')
print('Recall KNN =', knn_rec, '%', '\n')

cnf_matrix_nb = confusion_matrix(y_test, knn_pred)
np.set_printoptions(precision=2)

# Plot non-normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix_nb, classes=iris.target_names,
                      title='KNN\'s Confusion matrix, without normalization')

# Plot normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix_nb, classes=iris.target_names, normalize=True,
                      title='KNN\'s Normalized confusion matrix')

plt.show()
```

## Neural Network MLP

```python
iris_neural = neur.fit(X_train,y_train)
neur_pred = neur.predict(X_test)

neur_acc = metrics.accuracy_score(y_test, neur_pred) * 100
neur_prec = metrics.precision_score(y_test, neur_pred, average='macro') * 100
neur_rec = metrics.recall_score(y_test, neur_pred, average='macro') * 100
print('Akurasi MLP =', neur_acc, '%')
print('Presisi MLP =', neur_prec, '%')
print('Recall MLP =', neur_rec, '%', '\n')

cnf_matrix_nb = confusion_matrix(y_test, neur_pred)
np.set_printoptions(precision=2)

# Plot non-normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix_nb, classes=iris.target_names,
                      title='MLP\'s Confusion matrix, without normalization')

# Plot normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix_nb, classes=iris.target_names, normalize=True,
                      title='MLP\'s Normalized confusion matrix')

plt.show()
```

# 3. Pembelajaran dengan 10-fold cross Validation

```python
kf = model_selection.KFold(n_splits=10)

def x_fold_cross_validation(model):
    i = 1;
    for train_index, test_index in kf.split(iris.data):
        print('FOLD', i)
        X_train, X_test = iris.data[train_index], iris.data[test_index]
        y_train, y_test = iris.target[train_index], iris.target[test_index]
        
        model_fit = model.fit(X_train, y_train)
        model_pred = model_fit.predict(X_test)
        
        model_acc = metrics.accuracy_score(y_test, model_pred) * 100
        model_prec = metrics.precision_score(y_test, model_pred, average='macro') * 100
        model_rec = metrics.recall_score(y_test, model_pred, average='macro') * 100
        print('Akurasi =', model_acc, '%')
        print('Presisi =', model_prec, '%')
        print('Recall =', model_rec, '%', '\n')
        
        i = i + 1
```

## Naive Bayes

```python
print('Pembelajaran 10-Fold Cross Validation menggunakan Naive Bayes\n')
x_fold_cross_validation(gnb)
```

## Decision Tree

```python
print('Pembelajaran 10-Fold Cross Validation menggunakan Decision Tree Learning\n')
x_fold_cross_validation(dtl)
```

## k-Nearest Neighbors

```python
print('Pembelajaran 10-Fold Cross Validation menggunakan KNN\n')
knn = neighbors.nearest_centroid.NearestCentroid()
x_fold_cross_validation(knn)
```

## Neural Network MLP

```python
print('Pembelajaran 10-Fold Cross Validation menggunakan Neural Network MLP\n')
x_fold_cross_validation(neur)
```
